### Задача 1: предобработка текста

#### Цель:
Научиться применять базовые методы предобработки текста с использованием библиотек NLTK и SpaCy

#### Требования:
1. Выполнить чистку данных
2. Выполнить удаление стоп-слов
3. Привести текст к нижнему регистру
4. Произвести лемматизацию

Инструкция содержит плейсходеры в скобках `< >`, которые вы должны заменить на код

Прежде чем начать предобработку текста, подготовимся: загрузим инструменты (библиотеки) и сам текст.

```python
# Установите необходимые библиотеки (замените плейсхолдеры на ваш код)
# Для NLTK:
import nltk

# Для SpaCy:
import spacy

import re
```

```python
# Загрузка текста с помощью командной строки
!curl -o data.txt https://raw.githubusercontent.com/vifirsanova/compling/main/tasks/task1/data.txt
```

```python
# Код для открытия файла
with open('data.txt', 'r', encoding='utf-8') as file:
  text = file.read()
```

### Этап 1: Чистка текста
Здесь мы удаляем пунктуацию, чтобы избавиться от ненужных символов.

```python
# Код для удаления пунктуации
cleaned_text = re.sub(r'[^\w\s]', '', text) #любой символ, который не является буквой, цифрой или пробелом
print("Текст без пунктуации:", cleaned_text)
```

### Этап 2: Приведение текста к нижнему регистру
Мы приводим текст к **единому** виду, чтобы учитывать вхождения одного и того же слова как **один тип** независимо от правописания (например, "Tittle" и "title")

```python
# Код для приведения к нижнему регистру
cleaned_text = cleaned_text.lower()
print("Текст в нижнем регистре:", cleaned_text)
```

### Этап 3: Удаление стоп-слов
На этом этапе очищаем текст от наиболее частотных слов, которые частотны во всех текстах на русском языке (служебные слова, местоимения).

```python
# Код для удаления стоп-слов с помощью NLTK
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_words = [word for word in cleaned_text.split() if word not in stop_words]
print("Текст без стоп-слов:", filtered_words)
```

### Этап 4: Лемматизация текста
Приведем все слова в тексте к начальной форме, чтобы не учитывать формы слов.

```python
nlp = spacy.load('en_core_web_sm')  # Загрузка английской модели
doc = nlp(' '.join(filtered_words))  # Применение модели к отфильтрованным словам
lemmatized_words = [token.lemma_ for token in doc]  # Получение лемм
print("Лемматизированный текст:", lemmatized_words)
```

### Этап 5: Вывод итого текста и проверка

Посмотрим, что у нас получилось:


```python
final_text = ' '.join(lemmatized_words)
print("Итоговый текст:", final_text)
```
